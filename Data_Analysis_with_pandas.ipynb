# Importing necessary libraries 


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
## importing data and knowing the column and row number
```

```python
x=pd.read_csv('C:/Users/user/Desktop/Dataets/Sales_data_project.csv',encoding='unicode_escape')
x.shape
```

```python
## To get the idea of Data
```

```python
x.head()
```

```python
## some basic information regarding the dataset
x.info()
```

```python
#dropping unrelated columns
x.drop(['Status','unnamed1'],axis=1,inplace=True)
x.head()
```

```python
# Now checking for null values
x.isnull().sum()
```

```python
# Dropping null values and checking
x.dropna(inplace=True)
x.isnull().sum()
```

```python
x
```

```python
# changing datatype and checking
x['Amount']=x['Amount'].astype('int64')
x['Amount'].dtype
```

```python
# To make a list of all column
x.axes[1].to_list()
```

```python
# To rename columns and checking
x=x.rename(columns={'Product_Category':'product_type','Cust_name':'customer_name'})
x.axes[1].to_list()
```

```python
# Some basic statistical analysis
x.describe()
```

```python
sns.countplot(x='Gender',data=x)
```

```python
x.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
Sales_genderWise=x.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Gender',y='Amount',data=Sales_genderWise)
for i in ax.containers:
    ax.bar_label(i,)
```

```python
# From the above chart we see that most of the customers are female and their purchasing capability is higher than male.
```

```python
ax=sns.countplot(data=x, x='Age Group',hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
Sales_AgeWise=x.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
Sales_AgeWise
sns.barplot(x='Age Group',y='Amount',data=Sales_AgeWise)
```

```python
# From the above chart we see that most of the buyers are of the age group of 26-35
```

```python
ax=sns.countplot(data=x,x='Age Group',hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Most of the buyers are  of the age group of 26-35 femble.
```

```python
#Now total number of orders of the top ten states
Sales_statewise=x.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=Sales_statewise,x='State',y='Amount')
for i in ax.containers:
    ax.bar_label(i)

```

```python
# Most of buyers are of 'Uttar Pradesh','Maharastra'
```

```python
ax=sns.countplot(data=x,x='Marital_Status')
sns.set({'figure.figsize':(6,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
Sales_maritalstatuswise=x.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
Sales_maritalstatuswise
sns.set({'figure.figsize':(7,6)})
ax=sns.barplot(data=Sales_maritalstatuswise,x='Marital_Status',y='Amount',hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Most of the buyers are married women and they have higher buying power.
```

```python
sns.set({'figure.figsize':(20,5)})
ax=sns.countplot(data=x,x='Occupation')
for i in ax.containers:
    ax.bar_label(i)
```

```python
Sales_occupation=x.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set({'figure.figsize':(20,5)})
ax=sns.barplot(data=Sales_occupation,x='Occupation',y='Amount')
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_productIDwise=x.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set({'figure.figsize':(20,5)})
ax=sns.barplot(data=sales_productIDwise,x='Product_ID',y='Orders')
for i in ax.containers:
    ax.bar_label(i)
    
```

```python
# top 10 mostly sold products
fig1,ax1=plt.subplots(figsize=(12,7))
x.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
